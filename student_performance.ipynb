## Modelling

```python
data.columns
```

```python
X = data.drop('grade', axis =1)
y = data[['gender', 'place_of_residence', 'parent_marital_status','family_size','mode_of_transportation','elementary_education','higher_institution_location']]
```

```python
from sklearn.model_selection import train_test_split
x_train, xtest, ytrain, ytest = train_test_split(X,y, test_size=0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(x_train,ytrain)
```

```python
predictions = model.predict(xtest)
```

```python
from sklearn import metrics
```

```python
import pickle
file = open('ann_model.pkl', 'wb')
pickle.dump(model,file)
```

```python
import matplotlib.pyplot as plt
feat_importances =pd.Series(model.feature_importances_, index = x_train.columns)
feat_importances.nlargest(5).plot(kind='barh', figsize=(20,5))
plt.show()
```
